```python
tweets_df = pd.read_csv('/content/drive/MyDrive/CS425-NLP project/datasets/BERT-emotional classifier/emotion_tweets_github.csv')
# tweets_df = pd.read_csv('/content/drive/MyDrive/Acads/Y3 Sem1/CS425 NLC/CS425-NLP project/datasets/BERT-emotional classifier/emotion_tweets_github.csv')
tweets_df
```

```python
#Check frequency
tweets_df['sentiment'].value_counts()
```

```python
## Remove Handles

def clean_text(X):
    X = X.split()
    X_new = [x for x in X if not x.startswith("@")]
    return ' '.join(X_new)

tweets_df['text'] = tweets_df['content'].apply(clean_text)
tweets_df
```

```python
## drop empty rows in df
tweets_df = tweets_df.drop(tweets_df.index[[tweets_df[tweets_df['text'] == ''].index]])
```

## Relabel Tweets Dataset


| Label | Original Dataset | New Labelled Dataset |
| :---  |    :----:        |          ---:        | 
| 0 | happiness, love, fun, surprise, relief, enthusiasm | joy |
| 1 | sadness  | sadness | 
| 2 | anger, hate | anger | 
| 3 | worry | fear | 
| 4 | neutral, boredom, empty | neutral |



       

```python
## rename column name
tweets_df = pd.DataFrame(tweets_df, columns = ['text','sentiment'])
tweets_df = tweets_df.rename(columns={"sentiment": "label"})

## Relabel
tweets_df['label'] = tweets_df['label'].map({'happiness':0, 'love':0, 'fun':0, 'worry':3, 'empty':4, 'surprise':0, 'sadness':1, 'hate':2, 'anger': 2, 'boredom':4, 'relief': 0, 'enthusiasm':0, 'neutral':4})
```

```python
## Sample 50% of "Joy" data
tweets_df_joy = tweets_df.loc[tweets_df['label'] == 0]
tweets_df_joy = tweets_df_joy.sample(frac = 0.6)

tweets_df = tweets_df.loc[tweets_df['label'] != 0]

frames_tweets = [tweets_df, tweets_df_joy]
tweets_df = pd.concat(frames_tweets).reset_index(drop=True)
tweets_df
```

```python
#Check frequency
tweets_df['label'].value_counts()
```

```python
import string 

tweets_df['word_count'] = tweets_df['text'].str.split().str.len()

count = lambda l1,l2: sum([1 for x in l1 if x in l2])
tweets_df['punct_count'] = tweets_df['text'].apply(lambda s: count(s, set(string.punctuation)))

tweets_df['total_word_punct_count'] = tweets_df['punct_count'] + tweets_df['word_count']

tweets_df = tweets_df.drop(['punct_count','word_count'], axis=1)
tweets_df = tweets_df.loc[tweets_df['total_word_punct_count'] <= 128].reset_index(drop=True)

tweets_df
```

| 0 | joyful, content, grateful, proud, confident, nostalgic, trusting, caring, faithful, surprised, impressed, anticipating, hopeful, excited | joy |  |
| 1 |  sad, devastated, sentimental, lonely, disappointed, embarrassed    | sadness |  |
| 2 |  disgusted, ashamed, angry, furious, annoyed, jealous   | anger |  |

```python
df_empatheticDialogue = pd.DataFrame(empatheticDialogue_dataset, 
                          columns = ['conv_id', 'utterance_idx', 'context', 'prompt', 
                                     'speaker_idx', 'utterance', 'selfeval', 'tags'])
df_empatheticDialogue = df_empatheticDialogue[['utterance','context']]
df_empatheticDialogue['context'] = df_empatheticDialogue['context'].map(
    {'joyful':0, 'content':0, 'grateful':0, 'proud':0, 'confident':0, 'nostalgic':0,
     'trusting':0, 'caring':0, 'faithful':0, 
     'terrified':3, 'afraid':3, 'guilty':3, 'anxious':3, 'apprehensive':3,
     'surprised':0, 'embarrassed':1, 'impressed':0,
     'sad':1, 'devastated':1, 'sentimental':1, 'lonely':1, 'disappointed':1, 
     'disgusted':2, 'ashamed':2, 
     'angry':2, 'furious':2, 'annoyed':2, 'jealous':2,
     'anticipating':0, 'hopeful':0, 'prepared':4, 'excited':0 })
df_empatheticDialogue = df_empatheticDialogue.rename({'utterance': 'text', 'context': 'label'}, axis='columns')
df_empatheticDialogue
```
